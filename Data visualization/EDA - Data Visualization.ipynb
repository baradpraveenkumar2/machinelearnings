```python
# Step 1: Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2: Load the 'mpg' dataset from Seaborn
data = sns.load_dataset('mpg')
```

```python
# Step 3: Univariate Visualization
```

```python
sns.histplot(data['mpg'])
```

```python
# Univariate Histogram
plt.figure(figsize=(8, 5))
sns.histplot(data['mpg'], bins=20, kde=True)
plt.title('Histogram of MPG')
plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Univariate Pie Chart
plt.figure(figsize=(8, 5))
data['origin'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Pie Chart of Car Origins')
plt.ylabel('')
plt.show()
```

```python
# Univariate Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x=data['mpg'])
plt.title('Box Plot of MPG')
plt.xlabel('Miles per Gallon (MPG)')
plt.show()
```

```python
# Step 4: Bivariate Visualization
```

```python
# Bivariate Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='model_year', y='mpg', hue='origin', data=data,errorbar=None)
plt.title('Line Plot of MPG over Model Year (colored by Origin)')
plt.xlabel('Model Year')
plt.ylabel('Miles per Gallon (MPG)')
plt.legend(title='Origin')
plt.show()
```

```python
plt.scatter(data.weight, data.mpg)
```

```python
# Bivariate Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='weight', y='mpg', hue='origin', data=data)
plt.title('Scatter Plot of MPG vs Weight (colored by Origin)')
plt.xlabel('Weight')
plt.ylabel('Miles per Gallon (MPG)')
plt.legend(title='Origin')
plt.show()
```

```python
# Bivariate Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='origin', y='mpg', data=data)
plt.title('Box Plot of MPG by Car Origin')
plt.xlabel('Car Origin')
plt.ylabel('Miles per Gallon (MPG)')
plt.show()
```
